# United States Marriage Statistics (2005 - 2017)

## 1. Load libraries and data

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_path = '/kaggle/input/united-states-marriage-statistics/Week 30 - United States Marriage Status 2005 to 2017.csv'
df = pd.read_csv(df_path, index_col='index')
```

## 3. Data Visualisation

```python
df_with_total = df
```

```python
df = df[df["Metric"] != "Total"]
```

```python
plt.figure(figsize=(10, 6))
plt.title("")
sns.scatterplot(x=df['Age Group'], y=df['Estimated Population'], hue=df['Metric'])
plt.show()
```

```python
male_married_in_2008 = df.loc[(df['Gender']=='Male') & (df['Metric']=='Married') & (df['Year']==2008)]
male_married_in_2008
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=male_married_in_2008['Age Group'], y=male_married_in_2008['Estimated Population'])
plt.title("Married male in 2008")
plt.show()
```

```python
male_married_in_2017 = df.loc[(df['Gender']=='Male') & (df['Metric']=='Married') & (df['Year']==2017)]
male_married_in_2017
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=male_married_in_2017['Age Group'], y=male_married_in_2017['Estimated Population'])
plt.title("Married male in 2017")
plt.show()
```

```python
# divorces
male_divorced_in_2017 = df.loc[(df['Gender']=='Male') & (df['Metric']=='Divorced') & (df['Year']==2017)]
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=male_divorced_in_2017['Age Group'], y=male_divorced_in_2017['Estimated Population'])
plt.title("Divorced male in 2017")
plt.show()
```

```python
# divorces
female_divorced_in_2017 = df.loc[(df['Gender']=='Female') & (df['Metric']=='Divorced') & (df['Year']==2017)]
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=female_divorced_in_2017['Age Group'], y=female_divorced_in_2017['Estimated Population'])
plt.title("Divorced female in 2017")
plt.show()
```
